# build dataset

```python
products = [product_01, product_02, product_03]
```

```python
columns = ["product_id", "product_link", "username", "review"]
```

```python
product_ids = []
product_links = []
usernames = []
reviews = []
```

```python
for product in products:
    for review in product["reviews"]:
        product_ids.append(product["product_id"])
        product_links.append(product["product_link"])
        usernames.append(review["username"])
        reviews.append(review["comment"])
```

```python
df_reviews = pd.DataFrame(data={
    "product_id": product_ids,
    "product_link": product_links,
    "username": usernames,
    "review": reviews
})
```
